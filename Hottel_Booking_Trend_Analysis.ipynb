**Importing Required Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

**Data Loading**

```python
df = pd.read_csv("/hotel_bookings_data.csv")
```

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], format='%d/%m/%Y')
```

```python
df.drop(['company','agent'], axis = 1, inplace = True)
df.dropna(inplace = True)
```

```python
df.describe()
```

```python
df = df[df['adr']<5000]
```

**Data Analysis and Visualizations**

```python
canceled_percent = round(df['is_canceled'].value_counts(normalize=True)*100,2)
print(canceled_percent)

plt.figure(figsize=(5,5))
plt.title('Reservation Status Count')
plt.bar(['Not Canceled','Canceled'],df['is_canceled'].value_counts(), color = ['green','red'])
plt.show()
```

```python
plt.figure(figsize=(10,5))
fig1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels,_ = fig1.get_legend_handles_labels()
fig1.legend(title='Cancellation Status', labels=['Not Canceled', 'Canceled'], bbox_to_anchor=(1, 1))
plt.title('Reservation Status in different hotels',size = 20)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
plt.show()
```

```python
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')['adr'].mean()
city_hotel = city_hotel.groupby('reservation_status_date')['adr'].mean()
```

```python
plt.figure(figsize=(20,10))
plt.title('Average Daily Rate in City Hotel and Resort Hotel', fontsize=20)
plt.xlabel('Date')
plt.ylabel('Average Daily Rate')
plt.plot(resort_hotel.index, resort_hotel, label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel, label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(15,8))
fig2 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'bright')
legend_labels,_ = fig2.get_legend_handles_labels()
fig2.legend(title='Cancellation Status', labels=['Not Canceled', 'Canceled'], bbox_to_anchor=(1, 1))
plt.title('Reservation Monthly Status',size = 20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title('ADR per month', fontsize = 25)
sns.barplot(x = 'month', y = 'adr', data = df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index(),palette='coolwarm')
plt.show()
```

```python
canceled_data = df[df['is_canceled']==1]
top_10_country = canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(10,8))
plt.title('Top 10 Countries with Most Reservations Canceled')
plt.pie(top_10_country, labels = top_10_country.index, autopct='%.2f%%')
plt.show()
```

```python
df['market_segment'].value_counts(normalize= True)
```

```python
canceled_data['market_segment'].value_counts(normalize= True)
```

```python
cancelled_df_adr = canceled_data.groupby('reservation_status_date')['adr'].mean().reset_index()
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

not_cancelled_data = df[df['is_canceled']==0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')['adr'].mean().reset_index()
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(20,10))
plt.title('Average Daily Rate', fontsize=20)
plt.xlabel('Date')
plt.ylabel('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Canceled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Canceled')
plt.legend()
```

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20,10))
plt.title('Average Daily Rate', fontsize=20)
plt.xlabel('Date')
plt.ylabel('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Canceled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Canceled')
plt.legend()
```
